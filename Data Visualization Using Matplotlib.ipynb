```python
import matplotlib.pyplot as plt
```

#### Bar Graph

```python
X = [1,5,3,6,7]
Y = [10,40,50,70,20]
plt.title("Bar Graph",fontsize=15)
plt.xlabel("X - Axis",fontsize=12)
plt.ylabel("Y - Axis",fontsize=12)
plt.bar(X,Y,color=['red','blue'],label="Example 1")
plt.legend()
plt.show()
```

```python
X = ["A","B","C","D","E"]
Y = [10,40,50,70,20]
plt.title("Bar Graph",fontsize=15)
plt.xlabel("X - Axis",fontsize=12)
plt.ylabel("Y - Axis",fontsize=12)
plt.bar(X,Y,color=['red','blue'],label="Example 1")
plt.legend()
plt.show()
```

```python
X1 = [1,5,3,6,7]
Y1 = [10,40,50,70,20]
X2 = [4,9,8,2,1]
Y2 = [30,90,70,20,25]
plt.title("Bar Graph",fontsize=15)
plt.xlabel("X - Axis",fontsize=12)
plt.ylabel("Y - Axis",fontsize=12)
plt.bar(X1,Y1,color='red',label="Example 1",width=0.5)
plt.bar(X2,Y2,color='blue',label="Example 2",width=0.5)
plt.legend()
plt.show()
```

### Line Plot

```python
X1 = [1,5,3,6,7]
Y1 = [10,40,50,70,20]
X2 = [4,9,8,2,1]
Y2 = [30,90,70,20,25]
plt.title("Line Plot Graph",fontsize=15)
plt.xlabel("X - Axis",fontsize=12)
plt.ylabel("Y - Axis",fontsize=12)
plt.plot(X1,Y1,color='red',label="Example 1")
plt.plot(X2,Y2,color='blue',label="Example 2")
plt.legend(loc=4)
plt.show()
```

#### Scatter Plot

```python
X = [1,5,3,6,7,8,9,10]
Y = [10,40,50,70,20,56,78,54]
plt.figure(figsize=(10,8))
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("X - Axis",fontsize=12)
plt.ylabel("Y - Axis",fontsize=12)
plt.scatter(X,Y,label="Example 1",marker='*',color='red',s=300)
plt.legend()
plt.show()
```

### Pie Chart

```python
plt.figure(figsize=(10,8))
slices = [23,10,12,10,5]
activities =["Python","Android","Web","IOT","Robotics"]
cols = ['red','blue','green','magenta','yellow']
plt.pie(slices,labels=activities,colors=cols,autopct='%1.2f%%'
        ,explode=(0,0.2,0,0,0.4))
plt.legend(loc=4)
plt.show()
```
